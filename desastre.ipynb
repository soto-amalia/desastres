```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt


# 1) Ruta del archivo
DATA_PATH ="disasters.csv"  # CAMBIA ESTE NOMBRE si tu CSV se llama diferente


def cargar_dataset():
    print(" Cargando dataset desde:", DATA_PATH)
    df = pd.read_csv(DATA_PATH)
    print("Dataset cargado correctamente")
    return df
df= cargar_dataset()


```

```python
def inspeccion_basica(df: pd.DataFrame):
    print("\n游댳 Dimensiones del dataset (filas, columnas):")
    print(df.shape)

    print("\n游댳 Primeras 5 filas:")
    print(df.head())

    print("\n游댳 Nombres de columnas:")
    print(df.columns.tolist())

    print("\n游댳 Informaci칩n general (tipos de datos y nulos):")
    print(df.info())

    print("\n游댳 Conteo de valores nulos por columna (top 10):")
    print(df.isna().sum().sort_values(ascending=False).head(10))

inspeccion_basica(df)

```

```python
def pruebas_rapidas(df: pd.DataFrame):
    print("\n游댳 Columnas disponibles para an치lisis:")
    print(df.columns.tolist())

    # EJEMPLOS (ajusta los nombres cuando veas las columnas reales):
    if "Year" in df.columns:
        print("\n游늵 N칰mero de eventos por a침o:")
        print(df["Year"].value_counts().sort_index())

    if "Disaster_Type" in df.columns:
        print("\n游늵 Tipos de desastre m치s frecuentes:")
        print(df["Disaster_Type"].value_counts().head(10))



pruebas_rapidas(df)



```
